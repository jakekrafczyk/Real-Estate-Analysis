```
import pandas as pd
import numpy as np
from google.colab import files
upload = files.upload()
```

```
df = pd.read_csv('search_allegheny_1500to2000.csv',index_col=0)
print(df.shape)
df.head()
```

```
results_sub = df[df['2020_Status'].notnull()]
results_sub = results_sub[results_sub['2020_Status'] != 'ParcelIDNotValid']
print(results_sub.shape)
results_sub.head()
```

```
results_sub['Yr_Bought'] = results_sub.Yr_Bought.replace(np.NaN, '9999')
results_sub.head()
```

```
results_sub['Year'] = 0

for i in results_sub.index:
  results_sub.loc[i,'Year'] = int(results_sub.loc[i, 'Yr_Bought'][-4:])

results_sub.head()
```

```
results_sub.Year.value_counts()
```

```
results_sub = results_sub[results_sub.Year < 2000]
results_sub.shape
```

```
results_sub['Mkt_Value'] = results_sub['Mkt_Value'].replace('\$|,', '', regex=True)
results_sub['Mkt_Value'] = pd.to_numeric(results_sub['Mkt_Value'])
results_sub = results_sub[results_sub.Mkt_Value < 370000]
results_sub.shape
```

```
results_sub = results_sub[results_sub.Mkt_Value > 100000]
results_sub.shape
```

```
results_sub
```

```
res_only = results_sub[results_sub.Owner_Code != 'Corporation']
res_only.shape
```

```
res_only = res_only[res_only.Year > 1900]
print(res_only.shape)
res_only
```

```
res_only.to_csv('small_mkt_value2000.csv',index=False)
```
